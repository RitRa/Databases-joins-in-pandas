```python
import pandas as pd
```

```python
# make id column the index
car = pd.read_csv("https://raw.githubusercontent.com/ianmcloughlin/datasets/master/cars-db/car.csv", index_col=0)
car.head(10)
```

```python
# make id column the index
person = pd.read_csv("https://raw.githubusercontent.com/ianmcloughlin/datasets/master/cars-db/person.csv", index_col=0)
person.head(10)
```

```python
# make id column the index
county = pd.read_csv("https://raw.githubusercontent.com/ianmcloughlin/datasets/master/cars-db/county.csv", index_col=0)
county.head(10)
```

```python
pd.merge(car, person, left_on="OwnerId", right_on="ID", how="left")
```

```python
pd.merge(car, person, left_on="OwnerId", right_on="ID", how="left")[["Registration", "Name"]].head()
```

```python
tmp = pd.merge(car, person, left_on="OwnerId", right_on="ID", how="left")
pd.merge(tmp, county, left_on="Address", right_on="Name",how="left")
tmp
```
